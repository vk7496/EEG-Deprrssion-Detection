```python
import numpy as np
import matplotlib.pyplot as plt
import mne.time_frequency
import mne
import os

# Include the code to load and filter the data to ensure raw_filtered is defined
# Define the path to the copied file in /tmp
file_path = "/tmp/sujeto008-1.edf"

# Check if the file exists in /tmp before attempting to load
if os.path.exists(file_path):
    # Load the .edf file
    raw = mne.io.read_raw_edf(file_path, preload=True)

    # Apply a band-pass filter between 1 and 40 Hz
    raw_filtered = raw.copy().filter(l_freq=1., h_freq=40.)

    # --- Create Epochs and calculate PSD on epochs ---

    # You would typically define events here. For demonstration,
    # we'll create dummy events based on the raw data length.
    # Replace this with your actual event extraction if you have events.
    events = mne.make_fixed_length_events(raw_filtered, id=1, duration=1.0) # Example: create events every 1 second

    # Create epochs from the filtered raw data
    # Adjust tmin and tmax based on your epoch definition
    epochs = mne.Epochs(raw_filtered, events, tmin=0, tmax=1.0, preload=True)


    # Calculate Power Spectral Density (PSD) using a method for Epochs
    # Using psd_welch on Epochs object
    psds, freqs = mne.time_frequency.psd_welch(epochs, fmin=1, fmax=40, n_fft=256)

    # --- Rest of your code to calculate band powers and display ---

    # Define frequency bands
    bands = {
        'delta': (1, 4),
        'theta': (4, 8),
        'alpha': (8, 12),
        'beta': (12, 30),
        'gamma': (30, 40)
    }

    # Calculate mean power for each band per channel
    band_powers = {}
    for band, (fmin, fmax) in bands.items():
        # Find indices for the current band
        idx = np.logical_and(freqs >= fmin, freqs <= fmax)
        # Calculate the mean power over the frequency band and across epochs
        # psds shape is (n_epochs, n_channels, n_freqs)
        band_power = psds[:, :, idx].mean(axis=(0, 2)) # Average across epochs and frequencies
        band_powers[band] = band_power

    # Display the output
    for band in band_powers:
        print(f"\n💡 {band.upper()} Power per channel:\n{band_powers[band]}")

    # You can also plot the PSD of the epochs
    epochs.plot_psd(fmin=1, fmax=40)

else:
    print(f"❌ Error: The file was not found in the temporary directory: {file_path}")
```

```python
import mne
import numpy as np

def extract_bandpower_features(edf_path):
    raw = mne.io.read_raw_edf(edf_path, preload=True, verbose=False)
    raw.filter(1., 50., fir_design='firwin', verbose=False)

    data = raw.get_data()
    sfreq = raw.info['sfreq']

    psds, freqs = mne.time_frequency.psd_array_welch(data, sfreq=sfreq, fmin=1., fmax=50., n_fft=2048, verbose=False)
    psds_mean = psds.mean(axis=0)

    bands = {
        'delta': (1, 4),
        'theta': (4, 8),
        'alpha': (8, 13),
        'beta':  (13, 30),
        'gamma': (30, 50)
    }

    features = {}
    for band, (fmin, fmax) in bands.items():
        idx = np.logical_and(freqs >= fmin, freqs <= fmax)
        band_power = psds[:, idx].mean(axis=1)
        features[band] = band_power.mean()

    return features
```
